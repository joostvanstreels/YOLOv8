```python
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
```

#### Stap 1. Train- en testdata genereren

```python
# Keuze voor model maken

train_df = pd.read_csv('keypoints_train_df_YOLOv8m.csv', index_col=0)
test_df = pd.read_csv('keypoints_test_df_YOLOv8m.csv', index_col=0)

#train_df = pd.read_csv('keypoints_train_df_YOLOv8n.csv', index_col=0)
#test_df = pd.read_csv('keypoints_test_df_YOLOv8n.csv', index_col=0)
```

#### Stap 2. Model trainen


```python
X_train.head()
```

```python
y_train_pred = np.full_like(y_train, 5)

for i in range(0, len(y_train_pred)):
    if X_train['v00'].loc[i] < 0.8: y_train_pred[i] = 2
    elif X_train['A'].loc[i] < 0.0: y_train_pred[i] = 0
    elif X_train['B'].loc[i] < 0.15 and X_train['B'].loc[i] > 0.05 and X_train['C'].loc[i] < 0.15 and X_train['C'].loc[i] > 0.05: y_train_pred[i] = 0
    elif (X_train['B'].loc[i] < 0.15 and X_train['B'].loc[i] > 0.05) | (X_train['C'].loc[i] < 0.15 and X_train['C'].loc[i] > 0.05): y_train_pred[i] = 1
    elif X_train['F'].loc[i] < 0.05 and X_train['G'].loc[i] < 0.05: y_train_pred[i] = 3
    elif X_train['H'].loc[i] < 0.1 and X_train['I'].loc[i] < 0.1: y_train_pred[i] = 4
    elif X_train['J'].loc[i] > 0.008: y_train_pred[i] = 5
    else: y_train_pred[i] = 6
```

```python
y_test_pred = np.full_like(y_test, 5)

for i in range(0, len(y_test_pred)):
    if X_test['v00'].loc[i] < 0.8: y_test_pred[i] = 2
    elif X_test['A'].loc[i] < 0.0: y_test_pred[i] = 0
    elif X_test['B'].loc[i] < 0.15 and X_test['B'].loc[i] > 0.05 and X_test['C'].loc[i] < 0.15 and X_test['C'].loc[i] > 0.05: y_test_pred[i] = 0
    elif (X_test['B'].loc[i] < 0.15 and X_test['B'].loc[i] > 0.05) | (X_test['C'].loc[i] < 0.15 and X_test['C'].loc[i] > 0.05): y_test_pred[i] = 1
    elif X_test['F'].loc[i] < 0.05 and X_test['G'].loc[i] < 0.05: y_test_pred[i] = 3
    elif X_test['H'].loc[i] < 0.1 and X_test['I'].loc[i] < 0.1: y_test_pred[i] = 4
    elif X_test['J'].loc[i] > 0.008: y_test_pred[i] = 5
    else: y_test_pred[i] = 6
```
